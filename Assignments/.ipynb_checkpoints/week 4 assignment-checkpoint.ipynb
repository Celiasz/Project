```python
import geopandas as gpd

import matplotlib.pyplot as plt

import networkx as nx

import osmnx as ox

import contextily as ctx
```

I want to see the data I am working with aka the cooling centers.

We can see that these only account for about half of the data which is 113. I want to get a location for South El Monte.

I will be sampling from one location as a starting point. The address is 1556 Central Ave, South El Monte which is a local senior center. I will be using its latitute and longitude instead of its address. 

```python
South_El_Monte_Cooling_Center = (34.04692, -118.04657)
network_type = 'walk'
trip_times = [5, 10, 15, 20] 
meters_per_minute = 75 
```

```python
G = ox.graph_from_point(South_El_Monte_Cooling_Center, dist=1000, network_type=network_type)
```

Let's plot it. 

```python
fig, ax = ox.plot_graph(G)
```

Let's now put our info on web mercator and get the nodes. 

```python
G = ox.project_graph(G, to_crs='epsg:3857')
```

```python
gdf_nodes, gdf_edges = ox.graph_to_gdfs(G)
```

```python
gdf_nodes, gdf_edges = ox.graph_to_gdfs(G)
```

```python
gdf_nodes.plot(figsize=(12,12))
```

Now, let's get the edges. 

```python
gdf_edges.plot(figsize=(12,12))
```

Important, I need to calculate the network statistics. 

```python
stats = ox.basic_stats(G, circuity_dist='euclidean')
stats
```

To get the center, I need to figure out the centroid and calculate it. 

```python
minx, miny, maxx, maxy = gdf_nodes.geometry.total_bounds
print(minx)
print(miny)
print(maxx)
print(maxy)
```

```python
centroid_x = (maxx-minx)/2 + minx
centroid_y = (maxy-miny)/2 + miny
print(centroid_x)
print(centroid_y)
```

```python
center_node = ox.distance.nearest_nodes(G,Y=centroid_y,X=centroid_x)
print('The id for the nearest node is ' + str(center_node))
```

Here it is!

```python
gdf_nodes.loc[[center_node]]
```

Let's plot nodes with the centroid. 

```python
fig, ax = plt.subplots(figsize=(12,12))

gdf_edges.plot(ax=ax,
               linewidth=0.5,
               edgecolor='gainsboro', 
               zorder=10)

gdf_nodes.plot(ax=ax, 
               markersize=2, 
               zorder=20)

gdf_nodes.loc[[center_node]].plot(ax=ax,
                                  color='r', 
                                  zorder=30)

ax.axis('off')

ctx.add_basemap(ax,source=ctx.providers.CartoDB.Positron)
```

### Isochrome
Now, we are getting to the exiting part which is to determine the walking times and distances and determine the color it will assigned to. 

```python
gdf_edges['walk_time'] = gdf_edges['length']/meters_per_minute
```

```python
iso_colors = ox.plot.get_colors(n=len(trip_times), 
                                cmap='magma', 
                                start=0, 
                                return_hex=True)
print(trip_times)
print(iso_colors)
```

```python
trip_times.sort(reverse=True)
print(trip_times)
print(iso_colors)
```

```python
time_color = list(zip(trip_times, iso_colors))
time_color
```

```python
for time,color in list(time_color):
    print('The color for '+str(time)+' minutes is ' + color)
```

```python
for time, color in list(time_color):

    subgraph = nx.ego_graph(G, center_node, radius=time, distance='time')

    print('There are ' + str(len(subgraph.nodes())) + ' nodes within ' + str(time) + ' minutes ')
    
    for node in subgraph.nodes():
        gdf_nodes.loc[node,'time'] = str(time) + ' mins'
        gdf_nodes.loc[node,'color'] = color
```

```python
gdf_nodes['color'].fillna('#cccccc', inplace=True)
```

```python
gdf_nodes.sample(5)
```

This is will allow us to see the distances by their color to determine which is the closest and farthest distance from the centroid. 

```python
gdf_nodes.plot(figsize=(12,12),
               color=gdf_nodes['color'],
               markersize=20)
```

```python
gdf_nodes[gdf_nodes['time']=='5 mins']
```

```python
gdf_5 = gdf_nodes[gdf_nodes['time']=='5 mins']
```

```python
gdf_5.plot(figsize=(12,12),color=gdf_5.color)
```

```python
gdf_5.unary_union.convex_hull
```

```python
isochrones = gdf_nodes.dissolve("time")
isochrones
```

I can now create polygons for walking distances. 

```python
isochrones = isochrones.convex_hull.reset_index()
isochrones
```

```python
isochrones.columns=['time','geometry']
```

There it is! 

```python
isochrones.plot(figsize=(10,10),alpha=0.2,cmap='inferno')
```

For my last trick, I will now add the basemap and centroid to make sure that it makes sense when deciding where exactly I'm talking about. 

```python
fig, ax = plt.subplots(figsize=(15,15))

isochrones.plot(alpha=0.4, 
                ax=ax, 
                column='time', 
                cmap='inferno', 
                legend=True,
                zorder=20)
 
gdf_nodes.loc[[center_node]].plot(ax=ax,color='r', zorder=30)

gdf_nodes.plot(ax=ax, 
               markersize=1, 
               zorder=10)

gdf_edges.plot(ax=ax,
               linewidth=0.5,
               alpha=0.2,
               zorder=10)

ax.axis('off')

ax.set_title('Walking areas from a South El Monte cooling center: ' + '1556 Central Ave, South El Monte, CA ')

ctx.add_basemap(ax,source=ctx.providers.CartoDB.Positron)
```
