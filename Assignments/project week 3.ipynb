## Normalize Data
Eighth, I am now going to normalize my data and focusing on my 3 populations. 

```python
gdf_no_pop = gdf[gdf['Total']==0]
```

```python
gdf['Percent Hispanic or Latino'] = gdf['Hispanic or Latino']/gdf['Total']*100
gdf['Percent American Indian and Alaska Native alone'] = gdf['American Indian and Alaska Native alone']/gdf['Total']*100
gdf['Percent Black or African American alone'] = gdf['Black or African American alone']/gdf['Total']*100
```

```python
gdf.sample(5)
```

Ninth, I am now comparing different populations to each other. I am focusing on 

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 12))


ax1, ax2 = axs

# percent Hispanic or Latino
gdf.plot(column='Percent Hispanic or Latino', 
            cmap='OrRd', 
            scheme='user_defined',
            classification_kwds={'bins':[25,50,75,100]},
            edgecolor='white', 
            linewidth=0., 
            alpha=0.75, 
            ax=ax1,
            legend=True
           )

ax1.axis("off")
ax1.set_title("Percent Hispanic or Latino")

# percent Black or African American alone
gdf.plot(column='Percent Black or African American alone', 
            cmap='OrRd', 
            scheme='user_defined',
            classification_kwds={'bins':[25,50,75,100]},
            edgecolor='white', 
            linewidth=0., 
            alpha=0.75, 
            ax=ax2,
            legend=True
           )

ax2.axis("off")
ax2.set_title("Percent Black or African American alone")
```

Tenth, I want to make sure I want to focus on places that have a population of more than 75 percent for each ethnic group. So I wil lbegin with Latinx, proceeded by Black folks. Native people will not be part of the equation as they do not have a population of more than 75 percent in the county. I will later analyze it with a lower percentage hold. I believe by focusing on areas that have a 75 percent threshold, it can foretell where the hotspots are and where there should be at least one cooling center.

Lastly, I can now add the basemap to the percentages of Black and Latinx populations. 

```python
gdf_web_mercator = gdf.to_crs(epsg=3857)
```

```python
# use subplots that make it easier to create multiple layered maps
fig, ax = plt.subplots(figsize=(15, 15))

# add the layer with ax=ax in the argument 
gdf_web_mercator[gdf_web_mercator['Percent Hispanic or Latino'] > 75].plot(ax=ax, alpha=0.8)

# turn the axis off
ax.axis('off')

# set a title
ax.set_title('Census Tracts with more than 75% Hispanic or Latino Population',fontsize=16)

# add a basemap
ctx.add_basemap(ax)
```

```python
# use subplots that make it easier to create multiple layered maps
fig, ax = plt.subplots(figsize=(15, 15))

# add the layer with ax=ax in the argument 
gdf_web_mercator[gdf_web_mercator['Percent Black or African American alone'] > 75].plot(ax=ax, alpha=0.8)

# turn the axis off
ax.axis('off')

# set a title
ax.set_title('Census Tracts with more than 75% Black or African American alone Population',fontsize=16)

# add a basemap
ctx.add_basemap(ax)
```
